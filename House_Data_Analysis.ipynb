```python
df = pd.read_csv("HouseData.csv")
df.head(40)
```

```python
pd.set_option("display.max_rows", None)   # show all rows
print(df.isnull().sum())
```

```python
#only rows with missing values
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
#replacing null values in numerical columns
df_num=df.select_dtypes(include=['float','int'])
df_num.columns
```

```python
for col in df_num.columns:
    df_num[col].fillna(df_num[col].mean(),inplace=True)

df[df_num.columns] = df_num
```

```python
# filled all null of the categorical column with their mode value 

for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.drop(columns=['Id', 'Alley', "PoolQC", 'Fence', 'MiscFeature'], inplace=True)
```

```python
#heatmap
corr = df.select_dtypes(include='int').corr()
sns.heatmap(corr)
```

Each cell shows the correlation coefficient (ranging from -1 to +1) between two numerical features.

Dark colors (close to 0) â†’ weak or no correlation

Bright colors (close to +1) â†’ strong positive correlation

Dark purple/black (close to -1) â†’ strong negative correlation

```python
corr['YrSold'].sort_values(ascending=False)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

mask_upper = np.tril(np.ones_like(corr, dtype=bool))
mask_lower = np.triu(np.ones_like(corr, dtype=bool))

sns.heatmap(corr, mask=mask_lower, cmap='magma', annot=False, ax=axes[1])
axes[1].set_title('Lower Half Heatmap')

plt.show()
```

```python
sns.heatmap(corr, mask=mask_upper, cmap='coolwarm', annot=False, ax=axes[0])
axes[0].set_title('Upper Half Heatmap')
```
