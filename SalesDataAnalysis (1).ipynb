# Multivariate Analysis

```python
num_data = df.select_dtypes(include=['float64'])
num_data
```

```python
corr = num_data.corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

Insights


*   No relationship between Postal Code and Sale. The correlation of -0.02 means postal codes (which represent locations) have no direct linear relationship with sales values in your dataset.
